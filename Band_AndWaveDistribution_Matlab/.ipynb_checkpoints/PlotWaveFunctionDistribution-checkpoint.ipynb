```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = 'Times New Roman'

Type = ['FullRelaxed']
Angle = ['0.0']#,'0.1','0.2','0.3','0.396','0.5','0.558','0.769','0.915','1.381','1.530','2.060','2.558']

#This part is for Each sublattice, Gamma point, conduct band edge
Ang = []
K_Wave_C1 = []
K_Wave_C2 = []
K_Wave_B = []
K_Wave_N = []

for ele in Angle:
    data_wave = np.genfromtxt("./plot_wave_K.dat")
    #data_wave_G = np.genfromtxt("./Ang"+ele+"/Band/"+Type[0]+"/plot_wave_Gamma.dat")
    print(ele,len(data_wave))
    Hsize = len(data_wave)
    data_pos = np.genfromtxt("./generate.xyz",skip_header=4)
    data_sub = np.genfromtxt("./sublatticesSorted.dat")

    pos_x = data_pos[:,1]
    pos_y = data_pos[:,2]
    pos_type = data_pos[:,0]

    pos_sub = data_sub[:,1]

    wave = data_wave[:,int(Hsize/2)]
    Ang.append(float(ele))

    final_x = []
    final_y = []
    final_w = []

    for (x,y,t,w,sub) in zip(pos_x,pos_y,pos_type,wave,pos_sub):
        if sub == 1:
            final_x.append(x)
            final_y.append(y)
            final_w.append(w*100)
            
    plt.figure()
    plt.scatter(final_x,final_y,c=final_w)
    plt.colorbar()
    plt.gca().set_aspect('equal',adjustable='box')
    plt.title('C1 at Valance Band Edge $|\Psi|^2$',fontsize=20)
    plt.xlabel('x ($\AA$)')
    plt.ylabel('y ($\AA$)')
    plt.savefig("Ang0.0_C1_v.jpg",dpi = 400)

    #print('C1 wave',sum(final_w))
    K_Wave_C1.append(sum(final_w))

    final_x = []
    final_y = []
    final_w = []

    for (x,y,t,w,sub) in zip(pos_x,pos_y,pos_type,wave,pos_sub):
        if sub == 2:
            final_x.append(x)
            final_y.append(y)
            final_w.append(w*100)
            
    plt.figure()
    plt.scatter(final_x,final_y,c=final_w,vmin=0.0,vmax=0.0010)
    plt.colorbar()
    plt.gca().set_aspect('equal',adjustable='box')
    plt.title('C2 at Valance Band Edge $|\Psi|^2$',fontsize=20)
    plt.xlabel('x ($\AA$)')
    plt.ylabel('y ($\AA$)')
    plt.savefig("Ang0.0_C2_v.jpg",dpi = 400)

    #print('C1 wave',sum(final_w))
    K_Wave_C2.append(sum(final_w))

    final_x = []
    final_y = []
    final_w = []

    for (x,y,t,w,sub) in zip(pos_x,pos_y,pos_type,wave,pos_sub):
        if t != 3 and sub == 3:
            final_x.append(x)
            final_y.append(y)
            final_w.append(w*100)
            
    plt.figure()
    plt.scatter(final_x,final_y,c=final_w,vmin=0.0,vmax=0.0010)
    plt.colorbar()
    plt.gca().set_aspect('equal',adjustable='box')
    plt.title('B at Valance Band Edge $|\Psi|^2$',fontsize=20)
    plt.xlabel('x ($\AA$)')
    plt.ylabel('y ($\AA$)')
    plt.savefig("Ang0.0_B_v.jpg",dpi = 400)

    #print('C1 wave',sum(final_w))
    K_Wave_B.append(sum(final_w))

    final_x = []
    final_y = []
    final_w = []

    for (x,y,t,w,sub) in zip(pos_x,pos_y,pos_type,wave,pos_sub):
        if t != 3 and sub == 4:
            final_x.append(x)
            final_y.append(y)
            final_w.append(w*100)
            
    plt.figure()
    plt.scatter(final_x,final_y,c=final_w,vmin=0.0,vmax=0.0012)
    plt.colorbar()
    plt.gca().set_aspect('equal',adjustable='box')
    plt.title('N at Valance Band Edge $|\Psi|^2$',fontsize=20)
    plt.xlabel('x ($\AA$)')
    plt.ylabel('y ($\AA$)')
    plt.savefig("Ang0.0_N_v.jpg",dpi = 400)

    #print('C1 wave',sum(final_w))
    K_Wave_N.append(sum(final_w))

f = open("Data_Full_conduct_K.dat","w")
f.write("Ang   C1   C2   B   N\n")
for (ang,C1,C2,B,N) in zip(Ang,K_Wave_C1,K_Wave_C2,K_Wave_B,K_Wave_N):
    print(ang,C1,C2,B,N)
    f.write(str(ang)+" "+str(C1)+" "+str(C2)+" "+str(B)+" "+str(N)+"\n")

f.close()
```
