```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt

plt.rcParams['axes.formatter.useoffset'] = False
# from scipy.interpolate import interp1d, splev, splrep
plt.rcParams['font.sans-serif'] = 'Times New Roman'
```

```python
def GetAngleCurve(ang,Type):
    x_vec = []
    y_vec = []
    theta1_v = []
    theta2_v = []
    theta3_v = []
    theta_ave_v = []
    f = open("FinalData/"+str(ang)+"_"+Type+".dat","r").readlines()
    for line in f:
        v = line.split()
        if len(v) == 5:
            x = float(v[0])
            y = float(v[1])
            theta1 = float(v[2])
            theta2 = float(v[3])
            theta3 = float(v[4])
            theta_ave = (theta1+theta2+theta3)/3.0
            if abs(theta_ave) < 0.8:
                x_vec.append(x)
                y_vec.append(y)
                theta1_v.append(theta1)
                theta2_v.append(theta2)
                theta3_v.append(theta3)
                theta_ave_v.append((theta1+theta2+theta3)/3.0)
                theta10 = theta1
                theta20 = theta2
                theta30 = theta3
                theta_ave0 = theta_ave
            else:
                x_vec.append(x)
                y_vec.append(y)
                theta1_v.append(theta10)
                theta2_v.append(theta20)
                theta3_v.append(theta30)
                theta_ave_v.append((theta10+theta20+theta30)/3.0)
    if ang in [1.530]:
        temp = '1.530'
    elif ang in [2.060]:
        temp = '2.060'
    else:
        temp = str(ang)
    Path = "../Ang"+temp+"/LAMMPS/"+Type+"/"
    g = open(Path+'generateG.xyz','r').readlines()[:4] # Get the unit cell size
    v = g[0].split()
    a1x = float(v[0])
    a1y = float(v[1])
    v = g[1].split()
    a2x = float(v[0])
    a2y = float(v[1])            
    
    plt.figure()
    im = plt.scatter(x_vec,y_vec,c=theta_ave_v,cmap="coolwarm",vmax=0.45,vmin=-0.45)
    plt.xlim(0,a1x*1.5)
    plt.ylim(0,a2y)
    plt.colorbar(im)
    plt.gca().set_aspect("equal")
    plt.savefig(str(ang)+"_"+Type+"_super.jpg",dpi=400)
    max_theta = max(theta_ave_v)
    min_theta = min(theta_ave_v)
    return [max_theta,min_theta]

def GetAngleCurve_NoFig(ang,Type):
    x_vec = []
    y_vec = []
    theta1_v = []
    theta2_v = []
    theta3_v = []
    theta_ave_v = []
    f = open("FinalData/"+str(ang)+"_"+Type+".dat","r").readlines()
    for line in f:
        v = line.split()
        if len(v) == 5:
            x = float(v[0])
            y = float(v[1])
            theta1 = float(v[2])
            theta2 = float(v[3])
            theta3 = float(v[4])
            theta_ave = (theta1+theta2+theta3)/3.0
            if abs(theta_ave) < 0.8:
                x_vec.append(x)
                y_vec.append(y)
                theta1_v.append(theta1)
                theta2_v.append(theta2)
                theta3_v.append(theta3)
                theta_ave_v.append((theta1+theta2+theta3)/3.0)
                theta10 = theta1
                theta20 = theta2
                theta30 = theta3
                theta_ave0 = theta_ave
            else:
                x_vec.append(x)
                y_vec.append(y)
                theta1_v.append(theta10)
                theta2_v.append(theta20)
                theta3_v.append(theta30)
                theta_ave_v.append((theta10+theta20+theta30)/3.0)
    max_theta = max(theta_ave_v)
    min_theta = min(theta_ave_v)
    return [max_theta,min_theta]
```

```python
def Temp_Function(ang,Type):
    Type = "FullRelaxed"
    x_vec = []
    y_vec = []
    theta1_v = []
    theta2_v = []
    theta3_v = []
    theta_ave_v = []
    f = open("FinalData/"+str(ang)+"_"+Type+".dat","r").readlines()
    for line in f:
        v = line.split()
        if len(v) == 5:
            x = float(v[0])
            y = float(v[1])
            theta1 = float(v[2])
            theta2 = float(v[3])
            theta3 = float(v[4])
            theta_ave = (theta1+theta2+theta3)/3.0
            if abs(theta_ave) < 0.8:
                x_vec.append(x)
                y_vec.append(y)
                theta1_v.append(theta1)
                theta2_v.append(theta2)
                theta3_v.append(theta3)
                theta_ave_v.append((theta1+theta2+theta3)/3.0)
                theta10 = theta1
                theta20 = theta2
                theta30 = theta3
                theta_ave0 = theta_ave
            else:
                x_vec.append(x)
                y_vec.append(y)
                theta1_v.append(theta10)
                theta2_v.append(theta20)
                theta3_v.append(theta30)
                theta_ave_v.append((theta10+theta20+theta30)/3.0)
    if ang in [1.530]:
        temp = '1.530'
    elif ang in [2.060]:
        temp = '2.060'
    else:
        temp = str(ang)
    Path = "../Ang"+temp+"/LAMMPS/"+Type+"/"
    g = open(Path+'generateG.xyz','r').readlines()[:4] # Get the unit cell size
    v = g[0].split()
    a1x = float(v[0])
    a1y = float(v[1])
    v = g[1].split()
    a2x = float(v[0])
    a2y = float(v[1]) 
    output1 = [a1x,a1y,a2x,a2y]
    output2 = [x_vec,y_vec,theta_ave_v]
    return output1,output2
```
